```python
def cone_search_queer():
    ra = str(float(input('ra'))*u.degree)
    dec = str(float(input('dec'))*u.degree)
    circ_unit = input('Degree -> Arcmin? \n (Y)es or (N)o')
    if circ_unit == 'Y' :
        rad_degree = float(input('Degree'))*u.degree
        rad_arcmin = rad_degree.to(u.arcmin) 
        cone_search_dimensions1_base = ("{ra},{dec},{rad_arcmin}")
        cone_search_dimensions1 = cone_search_dimensions1_base.format(ra=ra, dec=dec, rad_arcmin=rad_arcmin)
        return cone_search_dimensions1
    else:
        cone_search_dimensions2_base = ("{ra},{dec},{rad_degree}")
        cone_search_dimensions2 = cone_search_dimensions2_base.format(ra=ra, dec=dec, rad_degree=rad_degree)
        return cone_search_dimensions2
cone_search_queer()
```

```python
ra = 88.8*u.degree
dec = 7.4*u.degree
coord_icrs = SkyCoord(ra=ra, dec=dec, frame='icrs')
coord_icrs
```

```python
coord_galactic = coord_icrs.transform_to('galactic')
coord_galactic
```

```python
gd1_frame = GD1Koposov10()
gd1_frame
```

```python
coord_gd1 = coord_icrs.transform_to(gd1_frame)
coord_gd1
```

```python
ra_1=0*u.degree
dec_1=0*u.degree

gd1_loc = SkyCoord(frame=gd1_frame, phi1=0*u.degree,phi2=0*u.degree)
gd1_icrs = gd1_loc.transform_to('icrs')
gd1_icrs
```

```python
phi1_min=-55*u.degree
phi1_max=-45*u.degree
phi2_min=-8*u.degree
phi2_max=4*u.degree

def make_rectangle(x1,x2,y1,y2):
    xs=[x1,x1,x2,x2,x1]
    ys=[y1,y2,y2,y1,y1]
    return xs,ys
```

```python
phi1_rect,phi2_rect = make_rectangle(phi1_min,phi1_max,phi2_min,phi2_max)
phi1_rect
```

```python
corners = SkyCoord(phi1=phi1_rect,phi2=phi2_rect,frame=gd1_frame)
corners_icrs = corners.transform_to('icrs')
corners_icrs
```

```python
corners_list_str=corners_icrs.to_string()
corners_single_str=' '.join(corners_list_str)
corners_single_str.replace(' ', ', ')
corners_single_str
```

```python
def skycoord_to_string(skycoord):
    corners_list_str=skycoord.to_string()
    corners_single_str=' '.join(corners_list_str)
    return corners_single_str.replace(' ', ', ')
```

```python
sky_point_list=skycoord_to_string(corners_icrs)
sky_point_list
```

```python
columns = "source_id, ra, dec, pmra, pmdec, parallax"
query3_base = """SELECT
TOP 10 
{columns}
FROM gaiadr2.gaia_source
WHERE parallax < 1
AND bp_rp BETWEEN -0.75 AND  2"""

polygon_top10query_base = """SELECT
TOP 10
{columns}
FROM gaiadr2.gaia_source
WHERE parallax < 1
  AND bp_rp BETWEEN -0.75 AND 2 
  AND 1 = CONTAINS(POINT(ra, dec), 
                   POLYGON({sky_point_list}))
"""

polygon_top10query =polygon_top10query_base.format(columns=columns, sky_point_list=sky_point_list)
print(polygon_top10query)
```
