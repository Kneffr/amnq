```python
query2= """ SELECT
TOP 3000
source_id, ra, dec, pmra, pmdec, parallax
FROM gaiadr2.gaia_source
WHERE bp_rp > -0.75 AND bp_rp < 2
"""
job2= Gaia.launch_job_async(query2)
job2
```

```python
results2 = job2.get_results()
results2
```

```python
column_format = 'source_id, ra, dec, pmra, pmdec, parallax'
max_parallax = 'parallax < 0.75'
query3_base = """ SELECT
TOP 10 
{column_format}
FROM gaiadr2.gaia_source
WHERE {max_parallax}
  AND bp_rp BETWEEN -0.75 AND 2
"""
query3 = query3_base.format(column_format=column_format, max_parallax=max_parallax)
job3=Gaia.launch_job(query3)
print(query3)
```

```python
results3= job3.get_results()
results3
```
